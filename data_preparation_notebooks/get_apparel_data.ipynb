```python
import pandas as pd
import numpy as np
import requests
import os
```

```python
df = pd.read_csv("amazon_products.csv")
```

```python
# apparel_category_id = [43, 84, 87, 88, 89, 90, 91, 94, 95, 96, 97, 98, 110, 112, 113, 114, 116, 118, 120, 121, 122, 123, 264, 265]
apparel_category_id = [110, 112, 113, 114, 116, 118, 120, 121, 122, 123]
```

```python
apparel_df = df[df["category_id"].isin(apparel_category_id)]
apparel_df = apparel_df[apparel_df["price"] > 0]
apparel_df["category_id"].value_counts()
```

```python
# get 2000 rows of each category
apparel_df = apparel_df.groupby("category_id").head(2000)
apparel_df["category_id"].value_counts()
```

```python
# Download images
count = 0
error = 0
for i, row in apparel_df[15000:].iterrows():
    url = row["imgUrl"]
    try:
        response = requests.get(url)
        if response.status_code == 200:
            with open("apparel_images\\" + row["asin"] + ".jpg", "wb") as f:
                f.write(response.content)
                print("Downloaded", row["asin"])
                count += 1
        else:
            print("Error downloading", row["asin"])
            error += 1
    except:
        print("Error downloading", row["asin"])
        error += 1

    if count % 1000 == 0:
        with open("log.txt", "a") as f:
            f.write(
                "Downloaded "
                + str(count)
                + " images + with "
                + str(error)
                + " errors\n"
            )

print("Downloaded", count, "images")
```

```python
apparel_df["imgName"] = apparel_df["asin"] + ".jpg"
apparel_df = apparel_df[
    ["asin", "title", "imgName", "price", "category_id", "imgUrl", "productURL"]
]
apparel_df
apparel_df.to_csv("apparel_products.csv", index=False)
```
